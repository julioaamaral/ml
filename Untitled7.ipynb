```python
def resizeAndPad(img, size, padColor=0):
    h, w = img.shape[:2]
    sh, sw = size

    # interpolation method
    if h > sh or w > sw: # shrinking image
        interp = cv2.INTER_AREA
    else: # stretching image
        interp = cv2.INTER_CUBIC

    # aspect ratio of image
    aspect = w/h

    # compute scaling and pad sizing
    if aspect > 1: # horizontal image
        new_w = sw
        new_h = np.round(new_w/aspect).astype(int)
        pad_vert = (sh-new_h)/2
        pad_top, pad_bot = np.floor(pad_vert).astype(int), np.ceil(pad_vert).astype(int)
        pad_left, pad_right = 0, 0
    elif aspect < 1: # vertical image
        new_h = sh
        new_w = np.round(new_h*aspect).astype(int)
        pad_horz = (sw-new_w)/2
        pad_left, pad_right = np.floor(pad_horz).astype(int), np.ceil(pad_horz).astype(int)
        pad_top, pad_bot = 0, 0
    else: # square image
        new_h, new_w = sh, sw
        pad_left, pad_right, pad_top, pad_bot = 0, 0, 0, 0

    # set pad color
    if len(img.shape) is 3 and not isinstance(padColor, (list, tuple, np.ndarray)): # color image but only one color provided
        padColor = [padColor]*3

    # scale and pad
    scaled_img = cv2.resize(img, (new_w, new_h), interpolation=interp)
    scaled_img = cv2.copyMakeBorder(scaled_img, pad_top, pad_bot, pad_left, pad_right, borderType=cv2.BORDER_CONSTANT, value=padColor)

    return scaled_img
```

```python
#Decomposicao Wavelet dos Canais RGB
def getSignalRGB(path,size): 
    imagem = cv2.imread(path)
    img = resizeAndPad(imagem, (size,size), padColor=0)
    b,g,r = cv2.split(img)
    r =  np.sort(np.asarray((r/r.max()*1.0).flatten(),dtype='float32'))  
    g = np.sort(np.asarray((g/g.max()*1.0).flatten(),dtype='float32'))
    b =  np.sort(np.asarray((b/b.max()*1.0).flatten(),dtype='float32'))
    return [r,g,b]

def getImageRGB(path, size): 
    imagem = cv2.imread(path)
    img = resizeAndPad(imagem, (size,size), padColor=0)
    b,g,r = cv2.split(img)
    rgb_t = cv2.merge((r,g,b))
   # plt.imshow(rgb_t)
   # plt.show()
    r =  np.asarray((r/r.max()*1.0),dtype='float32')
    g = np.asarray((g/g.max()*1.0),dtype='float32')
    b =  np.asarray((b/b.max()*1.0),dtype='float32')
    rgb = cv2.merge((r,g,b))
    return [rgb,r,g,b]

def infoImage(channel_1,channel_2,channel_3):
    print("Channel_1:"+str(channel_1))
    print("dtype: "+str(channel_1.dtype))
    print("Channel_1 Data Sorted:")
    plt.plot(channel_1)
    print("Channel_2:"+str(channel_2))
    print("dtype: "+str(channel_2.dtype))
    print("Channel_2 Data Sorted:")
    plt.plot(channel_2)
    print("Channel_3:"+str(channel_3))
    print("dtype: "+str(channel_3.dtype))
    print("Channel_3 Data Sorted:")
    plt.plot(channel_3)
       
def rms(channel, axis=None):
    return sqrt(mean(channel**2, axis=axis))  

def dwt1d(sinal,wavelet):
    coeffs = pywt.dwt(sinal, wavelet, mode='periodization')
    return(coeffs)

def dwt2d(sinal,wavelet2d):
    coeffs = pywt.dwt2(sinal, wavelet2d)
    return(coeffs) 
```
