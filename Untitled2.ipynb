```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread("00038.jpg")
```

```python
plt.imshow(image)
plt.show()
print(image.shape)
```

```python
red = image[:,:,0]
plt.imshow(red)
plt.show()
#red.shape
#red
```

```python
green = image[:,:,1]
plt.imshow(green)
plt.show()
#green.shape
#green
```

```python
blue = image[:,:,2]
plt.imshow(blue)
plt.show()
#blue.shape
#blue
```

```python
#Wavelets 
```

```python
import pywt 
```

```python
#Decomposicao do Canal Vermelho
data_red = red
print(data_red.shape)

wavelet_mother = 'bior1.1'

coeffs1_r = pywt.dwt2(data_red, wavelet_mother, mode='periodization')
cA1_r, (cH1_r, cV1_r, cD1_r) = coeffs1_r
print(cA1_r.shape)
print(cH1_r.shape)
print(cV1_r.shape)
print(cD1_r.shape)


coeffs2_r = pywt.dwt2(cA1_r, wavelet_mother, mode='periodization')
cA2_r, (cH2_r, cV2_r, cD2_r) = coeffs2_r
print(cA2_r.shape)
print(cH2_r.shape)
print(cV2_r.shape)
print(cD2_r.shape)

coeffs3_r = pywt.dwt2(cA2_r, wavelet_mother, mode='periodization')
cA3_r, (cH3_r, cV3_r, cD3_r) = coeffs3_r
print(cA3_r.shape)
print(cH3_r.shape)
print(cV3_r.shape)
print(cD3_r.shape)
```

```python
#Decomposicao do Canal Verde
data_green = green
print(data_green.shape)

coeffs1_g = pywt.dwt2(data_green, wavelet_mother, mode='periodization')
cA1_g, (cH1_g, cV1_g, cD1_g) = coeffs1_g
print(cA1_g.shape)

coeffs2_g = pywt.dwt2(cA1_g, wavelet_mother, mode='periodization')
cA2_g, (cH2_g, cV2_g, cD2_g) = coeffs2_g
print(cA2_g.shape)

coeffs3_g = pywt.dwt2(cA2_g, wavelet_mother, mode='periodization')
cA3_g, (cH3_g, cV3_g, cD3_g) = coeffs3_g
print(cA3_g.shape)
```

```python
#Decomposicao do Canal Azul
data_blue = blue
print(data_blue.shape)

coeffs1_b = pywt.dwt2(data_blue, wavelet_mother, mode='periodization')
cA1_b, (cH1_b, cV1_b, cD1_b) = coeffs1_b
print(cA1_b.shape)

coeffs2_b = pywt.dwt2(cA1_b, wavelet_mother, mode='periodization')
cA2_b, (cH2_b, cV2_b, cD2_b) = coeffs2_b
print(cA2_b.shape)

coeffs3_b = pywt.dwt2(cA2_b, wavelet_mother, mode='periodization')
cA3_b, (cH3_b, cV3_b, cD3_b) = coeffs3_b
print(cA3_b.shape)
```
